# Lesson 1: Agentic RAG - Code Along 4 - Persistent RAG

Previously, we saved the search index to an SQLite DB.
Now use it from this notebook to demonstrate it works.

```python
# dependencies
from sqlitesearch import TextSearchIndex
from rag_helper import RAGBase
from anthropic import Anthropic
```

```python
# connect to db and load data
sqlite_index = TextSearchIndex(
    text_fields=["question", "section", "answer"],
    keyword_fields=["course"],
    db_path="faq.db"
)
```

```python
# check number of documents in there
# should be 83
sqlite_index.count()
```

```python
# try a search
results = sqlite_index.search("Can I still join the course after it started?", num_results=5)
[doc["question"] for doc in results]
```

```python
# do rag using the persistent index
anthropic_client = Anthropic()

assistant = RAGBase(
    index=sqlite_index,
    llm_client=anthropic_client,
)
```

```python
answer = assistant.rag("Can I still join the course after it started?")
print(answer)
```

```python
# close the index
# actually Python could also handle this when shutting down the kernel
sqlite_index.close()
```
