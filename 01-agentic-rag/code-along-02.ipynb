```python
# environment variables
from dotenv import load_dotenv

# model provider
from anthropic import Anthropic

# custom code
from ingest import load_faq_data, build_index
from rag_helper import RAGBase
```

```python
# load environment variables from .env file
load_dotenv()
```

```python
# prepare FAQ data as search index for RAG knowledge base
documents = load_faq_data()
index = build_index(documents)
```

```python
# initialize Anthropic client for LLM responses
anthropic_client = Anthropic()
```

```python
# initialize a RAG assistant using the index, Anthropic client and custom class
assistant = RAGBase(
    index=index,
    llm_client=anthropic_client,
)
```

```python
# ask a question to the assistant to see if it works
# without RAG, it should not know what we're talking about
# it should only be able to answer because we provide context
# while the course instructor is also know publicly, of course,
# without RAG it would not know which course we participate in
answer = assistant.rag("I joined late - can I still enroll?")
print(answer)
```
